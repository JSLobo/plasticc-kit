# Separación de las muestras 

Esta es un pequeño ejemplo de cómo usar el método creado para separar las series de tiempo que resultan de las diferentes observaciones de un objeto.

Primero importamos las bibliotecas tradicionales y además traemos la clase que creamos para separar las series de tiempo.

```python
import numpy as np
import pandas as pd
from separate_samples import Separate_samples
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import Set1
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool
from collections import OrderedDict
output_notebook()
```

Cargamos los datos

```python
df_train = pd.read_csv('data/fake010.csv')
obj_10 = df_train[df_train['object_id'] == 10]
```

E instanciamos la clase, que nos permitirá separar las series de tiempo del objeto 615. Sólo se debe pasar el objeto. El resultado es una lista que contiene las series resultantes como DataFrames

```python
ss = Separate_samples()

samples = ss.get_separated_samples(obj_10)
samples[0]
```

```python
obj = samples[0]


#passband = [0,1,2,3,4,5]
#color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
#for band in passband:
#    pass_1 = obj_713[obj_713['passband'] == band]
#    x = pass_1['mjd']
#    y = pass_1['flux']
#    plt.subplot(6,1,band+1)
#    plt.xlabel('Time')
#    plt.plot(x,y,color[band])
#    plt.title('Passband '+ str(band))
#    plt.grid(True)

TOOLS="crosshair,pan,wheel_zoom,box_zoom,reset,hover,previewsave"
###############

passband = [0,1,2,3,4,5]
p = figure(plot_width=800, plot_height=800, tools=TOOLS)

for band in passband:
    passb = obj[obj['passband'] == band]
    x = passb['mjd']
    y = passb['flux']
    p.line(x, y, color=Set1[6][band])
    p.circle(x, y, color=Set1[6][band])
    

hover =p.select(dict(type=HoverTool))
hover.tooltips = OrderedDict([
    ("index", "$index"),
    ("(xx,yy)", "(@x, @y)"),
    ("label", "@label"),
])

show(p)
```
